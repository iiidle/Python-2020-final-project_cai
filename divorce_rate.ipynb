```python
import pandas as pd
import numpy as np
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import scale
```

```python
df_sum=pd.read_excel('df_sum2.xlsx')
```

```python
df_sum.drop('Unnamed: 0.1',axis=1,inplace=True)
```

```python
df_sum
```

```python
divorce=df_sum.pivot(index='reg',columns='Year',values='divorce_rate')
```

```python
divorce.dropna(axis=1,inplace=True)
```

```python
divorce
```

```python
pcafit=PCA(n_components=8)
```

```python
PCA_div=pcafit.fit_transform(divorce)
```

```python
pcafit.explained_variance_ratio_
```

```python
pcafit.explained_variance_
```

```python
pcafit2=PCA(n_components=1)
```

```python
len(pcafit.components_)
```

```python
PCA_div=pcafit2.fit_transform(divorce)
```

```python
print(pcafit2.explained_variance_ratio_)
print(pcafit2.explained_variance_)
print(pcafit2.get_params)
```

```python

pcafit2.components_
```

```python
len(pcafit2.components_[0])
```

```python
pcafit2.singular_values_ 
```

```python
X_cov = pcafit2.get_covariance() 
print(X_cov)
```

```python
pcafit2.components_
```

```python
divorce['divorce_pca_score']=PCA_div
```

```python
divorce
```

```python
divorce.to_excel("divorce_cluster.xlsx")
```
